# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

Данные с оценками по фильмам скачано с Kaggle.com https://www.kaggle.com/rounakbanik/the-movies-dataset потому, что на сайте grouplens.org есть проблемы со скачиванием датасетов (ошибка 404). 

```python
ratings = pd.read_csv('ratings_small.csv')
ratings.head()
```

```python
ratings.info()
```

```python
movies_meta = pd.read_csv('movies_metadata.csv')
movies_meta.head()
```

```python
movies = movies_meta[['id', 'title']]
movies.head()
```

```python
movies.info()
```

```python
links = pd.read_csv('links_small.csv', 
                    dtype={'movieId': 'int64', 'imdbId': 'object', 'tmdbId': 'int64'})
links.head()
```

```python
links.info()
```

```python
rating_links= pd.merge(ratings, links, on='movieId', how='left')
```

```python
rating_links
```

```python
rating_links_mark_5 = rating_links[rating_links.rating == 5.0]
movie_stats = rating_links_mark_5[['tmdbId', 'rating']].groupby('tmdbId').count().reset_index().sort_values('rating', ascending=False)
movie_stats.columns = ['tmdbId', 'cnt_mark']
movie_stats = movie_stats.reset_index(drop=True)
rating_extend = pd.merge(movie_stats, movies, left_on='tmdbId', right_on='id', how='left')
print(f'{rating_extend.title.loc[0]} - фильм с самым большим количеством оценок 5.0')
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
data.info()
```

```python
data.country.unique()
```

```python
data_fltr = data[((data.country == 'Latvia') | (data.country == 'Lithuania') | (data.country == 'Estonia'))
     & ((data.category == 4) | (data.category == 12) | (data.category == 21))
     & ((2005 <= data.year) & (data.year <= 2010))
     & (data.quantity > 0)
    ]
# Суммарное потребление
data_fltr.quantity.sum()
```
