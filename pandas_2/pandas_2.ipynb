# Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

````python
movies = pd.read_csv(r'ml-latest-small\movies.csv')
movies.head()
````

````python
movies.info()
````

````python
ratings = pd.read_csv(r'ml-latest-small\ratings.csv')
ratings.head()
````

````python
ratings.info()
````

````python
ratings_movies = pd.merge(ratings, movies, how='inner', on='movieId')
ratings_movies.head(20)
````

````python
ratings_movies.info()
````

````python
class_movie = ratings_movies.groupby(['movieId', 'title']).mean().reset_index()[['movieId', 'title', 'rating']]
class_movie.head(10)
````

````python
class_movie.info()
````

````python
def classification_movies(mark):
    if mark <= 2.0:
        return 'низкий рейтинг'
    elif mark <= 4.0:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
````

````python
class_movie['class'] = class_movie['rating'].apply(classification_movies)
````

````python
class_movie.sample(10)
````

## Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```JSON
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

Результат классификации запишите в отдельный столбец region.

````python
keywords = pd.read_csv(r'ml-latest-small\keywords.csv')
keywords.head()
````

````python
keywords.info()
````

````python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
````

````python
GEO_DATA_F = {}
for key, value in geo_data.items():
    for city in value:
        GEO_DATA_F[city] = key
````

````python
GEO_DATA_F
````

````python
def geo_classification(keyword):
    cities = GEO_DATA_F.keys()
    words = keyword.lower().split(' ')
    for word in words:
        if word in cities:
            return GEO_DATA_F[word]
        
    return 'undefined'
````

````python
keywords['region'] = keywords['keyword'].apply(geo_classification)
keywords.sample(20)
````

````python
keywords['region'].value_counts()
````

# Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
    1. для каждой строки пройдите по всем годам списка years
    2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

````python
movie_rat = class_movie.copy(False)[['movieId', 'title', 'rating']]
movie_rat.info()
````

````python
years = [str(x) for x in range(1950, 2011)]
````

````python
def production_year(title):
    for year in years:
        if year in title:
            return year
    return '1900'
````

````python
movie_rat['year'] = movie_rat['title'].apply(production_year)
````

````python
movie_rat.sample(10)
````

````python
movie_rat['year'].value_counts()
````

````python
stat_by_year = movie_rat.groupby(['year']).mean().reset_index().sort_values(['rating'], 
                                                                            ascending=False)[['year', 
                                                                                              'rating']]
stat_by_year.head(30)
````
